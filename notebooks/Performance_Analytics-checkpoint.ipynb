```python
import pandas as pd

# Load NAV data
nav = pd.read_csv("data/processed/nav_cleaned.csv")

# Convert date
nav['date'] = pd.to_datetime(nav['date'])

# Sort
nav = nav.sort_values(['amfi_code', 'date'])

# Daily return
nav['daily_return'] = nav.groupby('amfi_code')['nav'].pct_change()

# Drop NA
nav = nav.dropna()

nav.head()
```

```python
nav['daily_return'].describe()
```

```python
import numpy as np

def calculate_cagr(df, years):
    df = df.sort_values('date')
    start = df.iloc[0]['nav']
    end = df.iloc[-1]['nav']
    return (end / start) ** (1/years) - 1

cagr_list = []

for code, group in nav.groupby('amfi_code'):
    if len(group) > 252*5:
        cagr_1 = calculate_cagr(group.tail(252), 1)
        cagr_3 = calculate_cagr(group.tail(252*3), 3)
        cagr_5 = calculate_cagr(group.tail(252*5), 5)

        cagr_list.append([code, cagr_1, cagr_3, cagr_5])

cagr_df = pd.DataFrame(cagr_list, columns=['amfi_code','cagr_1yr','cagr_3yr','cagr_5yr'])

cagr_df.head()
```

```python
rf = 0.065  # 6.5%

sharpe_list = []

for code, group in nav.groupby('amfi_code'):
    mean_return = group['daily_return'].mean()
    std_dev = group['daily_return'].std()

    sharpe = ((mean_return - rf/252) / std_dev) * (252**0.5)

    sharpe_list.append([code, sharpe])

sharpe_df = pd.DataFrame(sharpe_list, columns=['amfi_code','sharpe_ratio'])
sharpe_df.head()
```

```python
sortino_list = []

for code, group in nav.groupby('amfi_code'):
    downside = group[group['daily_return'] < 0]['daily_return']

    downside_std = downside.std()
    mean_return = group['daily_return'].mean()

    sortino = ((mean_return - rf/252) / downside_std) * (252**0.5)

    sortino_list.append([code, sortino])

sortino_df = pd.DataFrame(sortino_list, columns=['amfi_code','sortino_ratio'])
sharpe_df.head()
```

```python
from scipy.stats import linregress

alpha_beta_list = []

# Use HDFC Top 100 as benchmark proxy
benchmark = nav[nav['amfi_code'] == 125497][['date','daily_return']]
benchmark = benchmark.rename(columns={'daily_return': 'benchmark_return'})

for code, group in nav.groupby('amfi_code'):

    merged = pd.merge(group, benchmark, on='date')
    merged = merged.dropna()

    if len(merged) > 100:
        slope, intercept, _, _, _ = linregress(
            merged['benchmark_return'],
            merged['daily_return']
        )

        beta = slope
        alpha = intercept * 252

        alpha_beta_list.append([code, alpha, beta])

alpha_beta_df = pd.DataFrame(alpha_beta_list, columns=['amfi_code','alpha','beta'])

alpha_beta_df.head()
```

```python
drawdown_list = []

for code, group in nav.groupby('amfi_code'):
    
    group = group.sort_values('date')

    running_max = group['nav'].cummax()
    drawdown = (group['nav'] / running_max) - 1

    max_dd = drawdown.min()

    drawdown_list.append([code, max_dd])

dd_df = pd.DataFrame(drawdown_list, columns=['amfi_code','max_drawdown'])

dd_df.head()
```

```python
final = cagr_df.merge(sharpe_df, on='amfi_code') \
               .merge(alpha_beta_df, on='amfi_code') \
               .merge(dd_df, on='amfi_code')

# Ranking
final['return_rank'] = final['cagr_3yr'].rank(ascending=False)
final['sharpe_rank'] = final['sharpe_ratio'].rank(ascending=False)
final['alpha_rank'] = final['alpha'].rank(ascending=False)
final['dd_rank'] = final['max_drawdown'].rank(ascending=True)

# Final Score
final['score'] = (
    0.30 * final['return_rank'] +
    0.25 * final['sharpe_rank'] +
    0.20 * final['alpha_rank'] +
    0.15 * final['dd_rank']
)

# Best funds on top
final = final.sort_values('score')

final.head()
```

```python
final.to_csv("data/processed/fund_scorecard.csv", index=False)
alpha_beta_df.to_csv("data/processed/alpha_beta.csv", index=False)

print("✅ Files saved successfully!")
```

```python
import numpy as np

cagr_list = []

for code, group in nav.groupby('amfi_code'):
    
    group = group.sort_values('date')

    if len(group) > 252:  # at least 1 year
        
        # 1 year
        cagr_1 = (group['nav'].iloc[-1] / group['nav'].iloc[-252]) ** (1/1) - 1
        
        # 3 year (if available)
        if len(group) > 252*3:
            cagr_3 = (group['nav'].iloc[-1] / group['nav'].iloc[-252*3]) ** (1/3) - 1
        else:
            cagr_3 = np.nan
        
        # 5 year (if available)
        if len(group) > 252*5:
            cagr_5 = (group['nav'].iloc[-1] / group['nav'].iloc[-252*5]) ** (1/5) - 1
        else:
            cagr_5 = np.nan

        cagr_list.append([code, cagr_1, cagr_3, cagr_5])

cagr_df = pd.DataFrame(cagr_list, columns=['amfi_code','cagr_1yr','cagr_3yr','cagr_5yr'])

cagr_df.head()
```
